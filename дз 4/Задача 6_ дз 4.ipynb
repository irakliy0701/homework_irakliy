# 1) n = 100

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
n = 100
u = np.random.normal(0,1,n)
d = np.random.normal(0,1,n)
```

```python
u @ u.T # Хотим получить матрицу, а не число !!!
```

```python
u = u[:, np.newaxis] # Делаем из u - вектор столбец. Чтобы посчитать u @ u.T (должна получиться матрица, а не число)
```

```python
u.shape
```

```python
# Cоберем матрицу А
A = (u @ u.T)/(u.T @ u) + np.diag(d)
np.min(LA.eigvalsh(A)) # минимальное собственное число
```

```python
sys = LA.eigh(A)
sys[0]
```

```python
sys[1]
```

```python
a = sys[:][:][1]
```

```python
c =a[:][0]
c # собственный вектор, соответствующий минимальному собственному занчению
```

# 2) n = 10**5

```python
n = 100000
d = np.random.normal(0, 1, (n,1))
u = np.random.normal(0,1, (n,1))
u = u/LA.norm(u)
```

```python
d_inv = 1/d
mu = 1 + u.T@(d_inv*u)
print(mu)
```

```python
####Ищем максимальное собственное число квадрата обратной матрицы
N = 0
x = np.zeros((n,1))
x[0,0] = 1 
h = np.zeros((n,1))
L_prev = 1
for i in range (1000):
    y = d_inv*x - (u.T@(d_inv*x))*(d_inv*u)/mu
    y = d_inv*y - (u.T@(d_inv*y))*(d_inv*u)/mu
    L = float(x.T@y)
    x = y/LA.norm(y)
    L_prev = L 
    if (np.abs(L - L_prev) < 10.**(-3)):
        h=x
        L = L_prev
h=x
print(LA.norm(h))
print(h.T, L)
print('\n')

x = np.zeros((n,1))
x[0,0] = 1
for i in range (100):
    y = d_inv*x - (u.T@(d_inv*x))*(d_inv*u)/mu
    y = d_inv*y - (u.T@(d_inv*y))*(d_inv*u)/mu
    L = float(x.T@y)
    N += 1
    x = y/LA.norm(y)
 
    print("Результат итерации "+f"{N}" +"-" + f"{x.T}" )
    print("Почти собственное число" +"  " + f"{L}")
    print("Невязка  " + f"{LA.norm(x-h).round(8)}")
    print('\n')
   
    if (LA.norm(x-h)<10.**(-8)):
        break

#Lambda = (h.T@C@h)/(h.T@h)
#print(Lambda)
```

```python
##### Проверяем, какое из корней найденного числа подходит в качестве собственного для просто обратной матрицы.
##### В данном примере получаем, что для отрицательного
La = np.sqrt(L_prev)
h_1 = d_inv*h - (u.T@(d_inv*h))*(d_inv*u)/mu
print(LA.norm((h_1 - La*h)))
print(LA.norm((h_1 + La*h)))

print(La)
```
