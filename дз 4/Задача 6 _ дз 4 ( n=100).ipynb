```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
n = 100
u = np.random.normal(0,1,n)
d = np.random.normal(0,1,n)
```

```python
u @ u.T # Хотим получить матрицу, а не число !!!
```

```python
u = u[:, np.newaxis] # Делаем из u - вектор столбец. Чтобы посчитать u @ u.T (должна получиться матрица, а не число)
```

```python
u.shape
```

```python
# Cоберем матрицу А
A = (u @ u.T)/(u.T @ u) + np.diag(d)
np.min(LA.eigvalsh(A)) # минимальное собственное число
```

```python
sys = LA.eigh(A)
sys[0]
```

```python
sys[1]
```

```python
a = sys[:][:][1]
c =a[:][0]
c # собственный вектор, соответствующий минимальному собственному занчению
```
