```python
# Сначала проанализируем поведение подинтегральной функции в зависимости от n (альфы берем как в задаче 0.1 и 10)
```

```python
# 1)График подинтегральной функции для а = 0.1

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
for n in range(1,30):
    x = np.linspace(0,1)
    y = (x**n)/(x + 0.1)
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
```

```python
# Получается, что при n -> беск. : интеграл (плошадь под графиком) -> 0
```

```python
# Для n = 25
x = np.linspace(0,1)
y = (x**25)/(x + 0.1)
plt.plot(x, y, '--', color ='orange')
```

```python
# 2)График подинтегральной функции для а = 10

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
for n in range(1,30):
    x = np.linspace(0,1)
    y = (x**n)/(x + 10)
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
```

```python
# Также как и в 1) получается, что при n -> беск. : интеграл (плошадь под графиком) -> 0
```

```python
# Для n = 25
x = np.linspace(0,1)
y = (x**25)/(x + 10)
plt.plot(x, y)
```

```python
# 3) Сравним I_25(0.1) и I_25(10) (сравниваем площадь под графиком)

x = np.linspace(0,1)
y = (x**25)/(x + 10)
x1 = np.linspace(0,1)
y1 = (x**25)/(x + 0.1)
plt.plot(x, y, '-', x1, y1,'--')
```

```python
# Получаем, что I_25(0.1) > I_25(10) !!! Однако оба неотрицательные
# Вообше, при любом n интеграл будет неотрицательным !
```
